# SMOTE + ENN and SMOTE + Tomek Links

This is a brief research on combined use of SMOTE and ENN/Tomek Links to amplify the minority class and remove noisy observations that might be created. 

The toy dataset is used.

Different separability levels are used. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs, make_classification

from imblearn.under_sampling import EditedNearestNeighbours, TomekLinks
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN, SMOTETomek
```

### Create data with separability level sep=2

```python
def make_data(sep):
    
    # returns arrays
    X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_redundant=0,
                           n_clusters_per_class=1,
                           weights=[0.99],
                           class_sep=sep,
                           random_state=1)
    
    # trasform arrays into pandas df and series
    X = pd.DataFrame(X, columns =['VarA', 'VarB'])
    y = pd.Series(y)
    
    return X, y

X, y = make_data(sep=2)
```

```python
sns.scatterplot(
    data=X, x="VarA", y="VarB", hue=y, alpha=0.5
)

plt.title('Toy dataset')
plt.show()
```

```python
sm = SMOTE(
    sampling_strategy='auto',
    random_state=0,
    k_neighbors=5,
    n_jobs=4
)

X_sm, y_sm = sm.fit_resample(X, y)
```

```python
# need ENN  as argument of SMOTEENN
enn = EditedNearestNeighbours(
    sampling_strategy='auto',
    n_neighbors=3,
    kind_sel='all',
    n_jobs=4)


smenn = SMOTEENN(
    sampling_strategy='auto',
    random_state=0,
    smote=sm,
    enn=enn,
    n_jobs=4
)

X_smenn, y_smenn = smenn.fit_resample(X, y)
```

```python
# need tomek as argument of SMOTETomek
tl = TomekLinks(
    sampling_strategy='all',
    n_jobs=4)

smtomek = SMOTETomek(
    sampling_strategy='auto', 
    random_state=0,
    smote=sm,
    tomek=tl,
    n_jobs=4
)

X_smtl, y_smtl = smtomek.fit_resample(X, y)
```

```python
# size of oversampled datasets

X_sm.shape, X_smenn.shape, X_smtl.shape
```

```python
# number of minority class observations

y.value_counts(), y_sm.value_counts(), y_smenn.value_counts(), y_smtl.value_counts()
```

```python
# plot of original data

sns.scatterplot(
    data=X, x="VarA", y="VarB", hue=y,alpha=0.5
)

plt.title('Original dataset')
plt.show()
```

```python
# plot of original data

sns.scatterplot(
    data=X_sm, x="VarA", y="VarB", hue=y_sm, alpha=0.5
)

plt.title('SMOTE dataset')
plt.show()
```

```python
sns.scatterplot(
    data=X_smenn, x="VarA", y="VarB", hue=y_smenn, alpha=0.5
)

plt.title('SMOTEENN dataset')
plt.show()
```

```python
sns.scatterplot(
    data=X_smtl, x="VarA", y="VarB", hue=y_smtl, alpha=0.5
)

plt.title('SMOTETomek dataset')
plt.show()
```

We can see that when the classes are not well separated, a combination of Over and undersampling can help remove the noisy samples created by SMOTE.

### data with sep=1

```python
def make_data(sep):
    
    # returns arrays
    X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_redundant=0,
                           n_clusters_per_class=1,
                           weights=[0.99],
                           class_sep=sep,
                           random_state=1)
    
    # trasform arrays into pandas df and series
    X = pd.DataFrame(X, columns =['VarA', 'VarB'])
    y = pd.Series(y)
    
    return X, y

X, y = make_data(sep=1)
```

```python
X_smenn, y_smenn = smenn.fit_resample(X, y)
X_smtl, y_smtl = smtomek.fit_resample(X, y)


# plot of original data with sep=1
sns.scatterplot(
    data=X, x="VarA", y="VarB", hue=y,alpha=0.5
)
plt.title('Original dataset')
plt.show()


# plot of SMOTEENN data with sep=1
sns.scatterplot(
    data=X_smenn, x="VarA", y="VarB", hue=y_smenn, alpha=0.5
)
plt.title('SMOTEENN dataset')
plt.show()

# plot of SMOTETomek data with sep=1
sns.scatterplot(
    data=X_smtl, x="VarA", y="VarB", hue=y_smtl, alpha=0.5
)
plt.title('SMOTETomek dataset')
plt.show()
```
